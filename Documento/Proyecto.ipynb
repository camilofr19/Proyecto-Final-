## Calcular Ventas Totales (2018-2021)

### Subtask:
Calcular las ventas totales de EV a partir de la `ev_sales_vehicles_df` DataFrame de los años 2018, 2019, 2020, and 2021.


```python
sales_2018_2021 = ev_sales_vehicles_df[ev_sales_vehicles_df['year'].isin([2018, 2019, 2020, 2021])]
total_ev_sales_2018_2021 = sales_2018_2021['value'].sum()
print(f"Total EV sales between 2018 and 2021: {total_ev_sales_2018_2021:,.0f} Vehicles")
```

## Identificar Año con Mayores Ventas

### Subtask:
Identificar el año con las ventas mas altas de EV sales del `ev_sales_vehicles_df` DataFrame.


```python
sales_by_year = ev_sales_vehicles_df.groupby('year')['value'].sum()
year_highest_sales = sales_by_year.idxmax()
max_sales_value = sales_by_year.max()

print(f"Year with the highest EV sales: {year_highest_sales}")
print(f"Highest EV sales in that year: {max_sales_value:,.0f} Vehicles")
```

## Identificar Región con Mayores Ventas

### Subtask:
Identificar la region con la mayor venta de EV del `ev_sales_vehicles_df` DataFrame.

```python
sales_by_region = ev_sales_vehicles_df.groupby('region')['value'].sum()
region_highest_sales = sales_by_region.idxmax()
max_sales_region_value = sales_by_region.max()

print(f"Region with the highest EV sales: {region_highest_sales}")
print(f"Highest EV sales in that region: {max_sales_region_value:,.0f} Vehicles")
```

## Identificar Región con Menores Ventas

### Subtask:
Identificar la región con las menores ventas de vehículos eléctricos del DataFrame `ev_sales_vehicles_df`.

```python
sales_by_region_min = ev_sales_vehicles_df.groupby('region')['value'].sum()
region_lowest_sales = sales_by_region_min.idxmin()
min_sales_region_value = sales_by_region_min.min()

print(f"Región con las menores ventas de vehículos eléctricos: {region_lowest_sales}")
print(f"Menores ventas de vehículos eléctricos en esa región: {min_sales_region_value:,.0f} Vehículos")
```

## Analizar Cuota de Mercado de Vehículos Eléctricos

### Subtask:
Filter the `datos` DataFrame to identify the 'EV sales share' data (where 'parameter' is 'EV sales share' and 'unit' is 'percent') and calculate the annual and regional trends for this market share.


```python
ev_sales_share_df = datos[(datos['parameter'] == 'EV sales share') & (datos['unit'] == 'percent')]

annual_sales_share_trend = ev_sales_share_df.groupby('year')['value'].mean()
regional_sales_share_trend = ev_sales_share_df.groupby('region')['value'].mean()

print("Annual EV Sales Share Trend (first 5 years):\n", annual_sales_share_trend.head())
print("\nRegional EV Sales Share Trend (first 5 regions):\n", regional_sales_share_trend.head())
```

## Analizar Ventas por Tipo de Tren Motriz

### Subtask:
Using the `ev_sales_vehicles_df` DataFrame, break down total sales by powertrain types: Battery Electric Vehicles (BEV) and Plug-in Hybrid Electric Vehicles (PHEV).


```python
powertrain_sales_df = ev_sales_vehicles_df[ev_sales_vehicles_df['powertrain'].isin(['BEV', 'PHEV'])]

annual_powertrain_sales = powertrain_sales_df.groupby(['year', 'powertrain'])['value'].sum().unstack()
regional_powertrain_sales = powertrain_sales_df.groupby(['region', 'powertrain'])['value'].sum().unstack()

print("Annual EV Sales by Powertrain (first 5 years):\n", annual_powertrain_sales.head())
print("\nRegional EV Sales by Powertrain (first 5 regions):\n", regional_powertrain_sales.head())
```

## Visualizar Tendencias de Cuota de Mercado

### Subtask:
Create a line plot showing the evolution of the 'EV sales share' globally and for relevant regions over time.


```python
plt.figure(figsize=(14, 7))

# Ver annual sales share trend (promedio global)
sns.lineplot(x=annual_sales_share_trend.index, y=annual_sales_share_trend.values, label='Global Average', marker='o', color='black')

# Filtrado regiones especificas para comparar
selected_regions = ['World', 'Europe', 'USA', 'China'] # Escoger regiones relevantes
regional_sales_share_filtered = ev_sales_share_df[ev_sales_share_df['region'].isin(selected_regions)]

# Grafico EV sales share de las regiones seleccionadas
sns.lineplot(data=regional_sales_share_filtered, x='year', y='value', hue='region', marker='o', palette='deep')

plt.title('Evolution of EV Sales Share Over Time (Global and Selected Regions)')
plt.xlabel('Year')
plt.ylabel('EV Sales Share (percent)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Region/Trend')
plt.tight_layout()
plt.show()
print("Line plot showing EV sales share evolution displayed.")
```

## Visualizar Desglose de Ventas por Tren Motriz

### Subtask:
Create a stacked bar chart illustrating the contribution of BEV and PHEV to total EV sales per year.


```python
plt.figure(figsize=(14, 7))
annual_powertrain_sales.plot(kind='bar', stacked=True, figsize=(14, 7), color=['#4CAF50', '#FFC107'], ax=plt.gca())
plt.title('Annual EV Sales by Powertrain Type (BEV vs. PHEV)')
plt.xlabel('Year')
plt.ylabel('Total EV Sales (Vehicles)')
plt.xticks(rotation=45)
plt.legend(title='Powertrain')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
print("Stacked bar chart showing annual EV sales by powertrain type displayed.")
```
